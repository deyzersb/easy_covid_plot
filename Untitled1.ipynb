```
country='Chile'
import pandas as pd
from datetime import datetime
def fecha_ahora():
  now = datetime.now()
  mes=now.month
  if (mes<10):
    mes='0'+str(mes)
  else:
    mes=str(mes)
  fecha=str(now.year)+"-"+mes+"-"+str(now.day)
  return fecha
data = pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv')
subdata = data[data['location']==country]
data_chile = subdata.set_index('date')
def covid_plot(parametro):
  print(parametro+" desde 2020-01-01 al "+fecha_ahora())
  if(parametro=="vacunas"):
    parametro='new_vaccinations_smoothed'
  if(parametro=="casos"):
    parametro='new_cases_smoothed'
  data_chile_1=data_chile.loc["2020-01-01":fecha_ahora(), parametro]
  data_chile_1.plot()
  #new_vaccinations_smoothed_per_million
```

```
covid_plot('new_vaccinations_smoothed_per_million')
```

```
covid_plot("casos")
```
